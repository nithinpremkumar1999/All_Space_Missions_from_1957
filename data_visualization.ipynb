```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
space_data = pd.read_csv('cleaned_data.csv')
space_data
```

```python
#Setting variable for categories in cleaned_data.csv
companies = space_data['Company Name']
countries = space_data['Country']
year = space_data['Year']
locations = space_data['Location']
rocket_status = space_data['Status Rocket']
mission_status = space_data['Status Mission']
```

```python
#Space companies vs Number of launches
plt.figure(figsize=(8, 20))
ax = sns.countplot(y="Company Name", data=space_data, order=companies.value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Company vs # of launches", fontsize=15)
ax.set_xlabel("Number of launches (log scale)", fontsize=13)
ax.set_ylabel("Company Name", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

```python
#Country vs Number of launches
plt.figure(figsize=(8, 20))
ax = sns.countplot(y="Country", data=space_data, order=countries.value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Country vs #launches", fontsize=15)
ax.set_xlabel("Number of launches(log scale)", fontsize=13)
ax.set_ylabel("Country", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

```python
#Locations vs Number of launches
plt.figure(figsize=(15,30))
ax = sns.countplot(y="Location", data=space_data, order=locations.value_counts().index)
ax.set_xscale("log")
ax.axes.set_title("Locations vs #launches", fontsize=15)
ax.set_xlabel("Number of launches(log scale)", fontsize=13)
ax.set_ylabel("Locations", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

```python
#Year vs Number of launches
plt.figure(figsize=(8, 20))
ax = sns.countplot(y="Year", data=space_data)
ax.axes.set_title("Year vs #Launches", fontsize=15)
ax.set_xlabel("Count", fontsize=13)
ax.set_ylabel("Year", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

```python
#Rocket status vs count
plt.figure(figsize=(8, 8))
ax = sns.countplot(y="Status Rocket", data=space_data, order=rocket_status.value_counts().index)
ax.axes.set_title("Status Rocket vs Count", fontsize=15)
ax.set_ylabel("Status Rocket", fontsize=13)
ax.set_xlabel("Count", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

```python
#Mission Status vs count
plt.figure(figsize=(10,10))
ax = sns.countplot(y="Status Mission", data=space_data, order=mission_status.value_counts().index)
ax.axes.set_title("Mission Status vs Count")
ax.set_xscale("log")
ax.set_xlabel("Count(log)", fontsize=13)
ax.set_ylabel("Mission Status", fontsize=13)
ax.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

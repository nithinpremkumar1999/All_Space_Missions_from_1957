```python
import pandas as pd
```

```python
raw_data = pd.read_csv('Space_corrected.csv')
raw_data
```

```python
#Dropping column Unnamed: 0
data = raw_data.copy()
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
#Dropping column Unnamed: 0.1
data = data.drop(['Unnamed: 0.1'], axis = 1)
```

```python
#Adding country column
data["Country"] = data["Location"].apply(lambda location: location.split(", ")[-1])
```

```python
#Changing Datum to pandas date time
data['DateTime'] = pd.to_datetime(data['Datum'])
data['Year'] = data['DateTime'].apply(lambda datetime: datetime.year)
```

```python
#Dropping DateTime column
data = data.drop('DateTime', axis=1)
```

```python
data
```

```python
data.to_csv(r'cleaned_data.csv', index=False)
```
